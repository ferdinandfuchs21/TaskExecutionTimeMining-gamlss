```python
import pm4py
import sys
import random

import pandas as pd
sys.path.append("../../TaskExecutionTimeMining")

from event_log_transformer import *
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
file_path = '../../../data/BPI Challenge 2017.xes'
event_log = pm4py.read_xes(file_path)
event_log
```

```python
event_log['lifecycle:transition'].unique()
```

```python
start_end_event_log2 = TransformEventLog.start_end_event_log_all(event_log)
start_end_event_log2 = TransformEventLog.seconds_in_day(start_end_event_log2, 'time:timestamp_start')
start_end_event_log2 = TransformEventLog.day_of_week(start_end_event_log2, 'time:timestamp_start')
start_end_event_log2
```

```python
start_end_event_log2 = start_end_event_log2[~((start_end_event_log2['lifecycle:transition_start'] == 'complete')
                      & (start_end_event_log2['lifecycle:transition_complete'] == 'complete'))]


start_end_event_log2[((start_end_event_log2['lifecycle:transition_start'] == 'suspend')
                      & (start_end_event_log2['lifecycle:transition_complete'] == 'suspend'))]

start_end_event_log2 = start_end_event_log2[~start_end_event_log2['case:concept:name'].isin(
    start_end_event_log2[start_end_event_log2['lifecycle:transition_start'] == start_end_event_log2['lifecycle:transition_complete']]['case:concept:name'].values
)]
```

```python
transformer = TransformEventLog()
start_end_event_log2 = transformer.add_cumulative_dummies(
    start_end_event_log2, 
    case_col='case:concept:name',
    cols_to_encode=['org:resource_start', 'concept:name']
)
```

```python
start_end_event_log2
```

```python
unique_concept_names = start_end_event_log2["concept:name"].unique()
print(unique_concept_names)
```

```python
org:resource_start_names = start_end_event_log2["org:resource_start"].unique()
print(resource_start_names)
```

```python
# Identify columns where all values are NaN
all_nan_cols = start_end_event_log2.columns[start_end_event_log2.isna().all()]
print("Columns with only NaN values:")
print(all_nan_cols.tolist())

start_end_event_log2 = start_end_event_log2.drop(columns=all_nan_cols)
```

```python
random.seed(42)

train, test = pm4py.ml.split_train_test(start_end_event_log2, case_id_key='case:concept:name')
```

```python
print(train.shape, test.shape)
print(train['case:concept:name'].unique().shape, test['case:concept:name'].unique().shape)
```

```python
train.to_csv('../../../data/BPIC2017/transformed_event_logs/BPIC2017_train.csv', index=False, date_format='%Y-%m-%d %H:%M:%S.%f')
train.to_pickle('../../../data/BPIC2017/transformed_event_logs/BPIC2017_train.pickle')
```

```python
test.to_csv('../../../data/BPIC2017/transformed_event_logs/BPIC2017_test.csv', index=False, date_format='%Y-%m-%d %H:%M:%S.%f')
test.to_pickle('../../../data/BPIC2017/transformed_event_logs/BPIC2017_test.pickle')
```
